# About Dataset  
Overview of the dataset:
* The medical insurance dataset contains information about a number of factors that can affect medical expenses, including age, sex, BMI, smoking status, number of children, and region. This dataset can be used to train a machine learning model that can predict medical expenses for new customers.

To provide insights into the key factors that contribute to higher insurance costs and help the company make more informed decisions regarding pricing and risk assessment.

The dataset contains 2.7K rows and 7 columns
Columns include:
1. Age
2. Sex
3. BMI (Body Mass Index)
4. Children
5. Smoker
6. Region
7. Charges

## Problem Statement:
* What are the most important factors that affect medical expenses?
* How well can machine learning models predict medical expenses?
* How can machine learning models be used to improve the efficiency and profitability of health insurance companies?

The data and problem statement was taken from [HARISH KUMARDATALAB on Kaggle Website](https://www.kaggle.com/datasets/harishkumardatalab/medical-insurance-price-prediction)

### Graphs  
Now we analyze some variables with graphs, to have a visual reference of the distribution of the data.

```python
#Create a 2X3 subplots
fig, ax = plt.subplots(nrows = 2, ncols = 3, figsize=(11, 8))

# Subplot [0,0] Age
sns.histplot(data = df["age"], ax = ax[0,0], color ='Orange')
ax[0,0].set_xlabel("Age")
ax[0,0].set_ylabel("Count")

#Subplot [0,1] Sex
labels = ["Male", "Female"]
counts = [df["sex"].value_counts()["male"], df["sex"].value_counts()["female"]]
ax[0,1].bar(labels, counts, color = ['blue', 'pink'])

#Subplot [0,2] Bmi
sns.histplot(x=df["bmi"], kde=True, ax=ax[0,2])
ax[0,2].set_ylabel("")

#Subplot [1,0] Children
sns.histplot(x=df['children'],color='skyblue', ax=ax[1,0])

#Subplot [1,1] Smoker
sns.countplot(x=df['smoker'], ax=ax[1,1])
ax[1,1].set_ylabel("")

#Subplot [1,2] Region
sns.countplot(x=df['region'], ax=ax[1,2])
ax[1,2].tick_params(axis='x', labelsize=8)
ax[1,2].set_ylabel("")

plt.show()
```

```python
#Create a 2X3 subplots
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize=(12, 5))

# Charges
#Histogram
sns.histplot(data=df["charges"], ax=ax[0])
ax[0].set_title('Charge Distribution')
ax[0].set_xlabel('Charges Amount')


#Box Plot
sns.boxplot(data=df["charges"], ax=ax[1], color = 'Orange')
ax[1].set_title('Charge BoxPlot')
ax[1].set_ylabel('Charges Amount')
```

## First Conclutions

**DataSet:** Have 2772 rows and 7 columns. We have 4 numerical variables (2 int / 2 float) and 3 categorical variables. No row contains null values, which saves us time by not having to work with NaN values.

**Age:** The data has a positive right skew, which indicates that the majority of the data tends to belong to young clients, between approximately 18 and 25 years old.  
**Sex:** The distribution is practically 50/50.  
**Bmi:** Have a normal distribution and a few outliers values in the right side.

**Smoker:** The majority of records are of non-smokers with more than 80%.  
**Region:** The distribution is very uniform with a small number of more records from southeast.  
**Charges:** The data has a positive very pronounced right skew, the 80% of records are between 1200 and 20200, and have a lot outliers in the right side.

## Encoding categorical variables
**Sex:** 0 = male / 1 = female  
**Smoker:** 0 = No Smoker / 1 = Smoker  
**Region:** I decided to use the get dummies function.

```python
df["sex"].replace({"male": 0, "female": 1}, inplace=True)
df["smoker"].replace({"no": 0, "yes": 1}, inplace=True)
region_labels = {"northeast":"ne", "northwest":"nw", "southeast":"se", "southwest":"sw"}
encoded_regions = pd.get_dummies(df["region"], dtype= int)
encoded_regions.rename(columns=region_labels, inplace = True)
df.drop(["region"],axis = 1, inplace = True)
df_encoded = pd.concat([df, encoded_regions], axis = 1)
df_encoded
```

```python
print(df_encoded.corr())
```

```python
sns.heatmap(df_encoded.corr().round(decimals=2), annot=True, cmap="magma",linewidths=0.5)
plt.show()
```
